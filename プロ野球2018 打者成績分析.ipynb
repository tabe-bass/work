pandasのread_htmlメソッドにより、Webページ内の表をデータフレームとして読み込む。
データはプロ野球Freak(https://baseball-data.com/)のものを使用。

```python
url = 'https://baseball-data.com/18/stats/hitter-all/tpa-2.html'
datasets = pd.io.html.read_html(url)
datasets[0]
```

htmlの表データからDataFrameを作成し、csvファイルに変換する

```python
fetched_dataframes = pd.io.html.read_html(url)
```
